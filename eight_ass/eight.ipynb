```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset("titanic")
df
```

```python
sns.displot(x=df['age'],bins=10,kde=False)
```

```python
sns.jointplot(x=df['age'],y=df['fare'],kind='scatter')
```

```python
sns.rugplot(df['fare'])
```

```python
sns.barplot(df['fare'])
```

```python
sns.barplot(x=df['sex'],y=df['age'],estimator=np.std)
```

```python
sns.countplot(x=df['sex'])
```

```python
sns.boxplot(x=df['sex'],y=df['age'],hue=df['survived'])
```

```python
sns.violinplot(x ='sex',y='age',hue='survived',data=df)
```

```python
sns.stripplot(x='sex',y='age',jitter=False,hue='survived',data=df)
```

```python
sns.stripplot(x='sex',y='age',jitter=True,hue='survived',data=df)
```

```python
sns.swarmplot(x='sex', y='age',hue='survived',data=df)
```
